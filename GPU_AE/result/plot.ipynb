```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
with open('CPU.txt', 'r') as f:
    lines = f.readlines()
    cpu = [float(line.strip()) for line in lines]

with open('2x_T4.txt', 'r') as f:
    lines = f.readlines()
    T4 = [float(line.strip()) for line in lines]

with open('GTX1650.txt', 'r') as f:
    lines = f.readlines()
    GTX1650 = [float(line.strip()) for line in lines]

with open('P100.txt', 'r') as f:
    lines = f.readlines()
    P100 = [float(line.strip()) for line in lines]
```

```python
# Plot lines plot in the same figure
plt.plot(cpu, label='CPU Intel(R) i5-10300H')
plt.plot(GTX1650, label='GPU NVIDIA GTX 1650')
plt.plot(T4, label='2x GPU T4')
plt.plot(P100, label='GPU P100')
plt.xlabel('Iteration #')
plt.ylabel('Execution Time (s)')
plt.xticks(np.arange(0, len(cpu), 1))
plt.yticks(np.arange(0, max(max(cpu), max(T4), max(GTX1650), max(P100)), 5))
plt.legend()
plt.title('Execution Time Comparison')
plt.show()
```
