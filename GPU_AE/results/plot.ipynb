```python
import matplotlib.pyplot as plt
import numpy as np

with open("tpr_parametric_plot.txt") as f:
    wdgrl_parametric = f.readlines()
wdgrl_parametric = [float(x.strip()) for x in wdgrl_parametric]

with open("tpr_oc_plot.txt") as f:
    wdgrl_oc = f.readlines()
wdgrl_oc = [float(x.strip()) for x in wdgrl_oc]
# Sample data
delta = np.array([1, 2, 3, 4])

# Plot the data
plt.plot(delta, wdgrl_parametric[:4], marker='o', linestyle='-', label="WDGRL-parametric", color='tab:blue')
plt.plot(delta, wdgrl_oc[:4], marker='o', linestyle='-', label="WDGRL-oc", color='tab:orange')

plt.xlabel("Delta")
plt.ylabel("TPR")
plt.xticks(delta)
plt.yticks([0.0, 0.2, 0.4, 0.6, 0.8, 1])  

# Legend
plt.legend()

# Show the plot
plt.show()
```
